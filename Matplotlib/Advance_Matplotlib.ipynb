```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Scatter Plot Colour



*   c-> this argument must be a color, a sequence of colors, or a  
sequence of numbers, not 0

*   cmap -> colour palatte names(check matplotlib documents)
*   alpha -> opacity

*   plt.colorbar()-> shows the color change in bar







```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# making the species column in numbers to persorm c argument
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
#plt.colorbar()

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(13,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
batters
```

```python
sample_df = batters.head(100).sample(25,random_state=5) #random_state -> That means every time you run the code with the same random_state and data, you'll get the exact same 25 samples.
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s = sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red') #axis horizontal line
plt.axhline(140,color='green')
plt.axvline(30,color='red') #axis vertical line

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Subplots

```python
# A diff way to plot graphs

batters.head()
```

```python
# this gives a tuple of two objects. one is "Figure Size" where we will plot the graphs and other is "Axes"

plt.subplots()
```

```python
fig,ax = plt.subplots(figsize=(15,6))  # fig,ax are variables to hold the two objects of plt.subplots() function

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')

ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show() #it is like plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1) # nrows/ncols set the number of subplots
ax
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True, figsize = (10,7)) #sharex -> if the x axis has same values then to synchronize this two graphs

ax[0].scatter(batters['avg'],batters['strike_rate'],color='orange')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# 4 graphs together
# for 4 graphs ax creates 2D object. So i have to give two inputs ax[x,y]

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'], color = 'orange')
ax[0,1].scatter(batters['avg'],batters['runs'],color = 'red')
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])


```

# 3D Surfaceplot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
# meshgrid(converts 1D array to 2D arrays)

m1,m2 = np.meshgrid(x,y)
```

```python
z = m1**2 + m2**2
z
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(m1) + np.cos(m2)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```

# Contour Plots

### It converts 3D plots to 2D Plots

```python
z = m1**2 + m2**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contourf -> fill the pattern in colour

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(m1) + np.cos(m2)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(m1) + np.cos(m2)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(m1,m2,z,cmap='viridis')
fig.colorbar(p)
```
