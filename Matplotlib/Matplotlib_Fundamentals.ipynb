### Hidden Parameters to style the graph


*   linestyle = solid/dashed/dotted/dashdot
*   line width
*   marker = d(diamond)/+(plus)/>(arrow)/.(dot)/o(big circle)
*   markersize

*   legend() and label -> create the box for labeling









```python
#line style (linestyle = solid/dashed/dotted/dashdot)
#line width

plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dotted',linewidth = 3,marker ='o',markersize=10,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color ='teal',linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli IPL Run Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

```

```python
# legend(loc =) parameters -> 'best',(default)
#                        'upper right',
#                        'upper left',
#                        'lower left',
#                        'lower right',
#                        'right',
#                        'center left',
#                        'center right',
#                        'lower center',
#                        'upper center',
#                        'center'


plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dotted',linewidth = 3,marker ='o',markersize=10,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color ='teal',linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli IPL Run Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color=colors)
```

```python
# barh() - horizontal bar chart

plt.barh(colors,children,color=colors)
```

```python
# bar chart using dataset
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

## Side by Side bar chart


```python
# Position of bars on x-axis
x = np.arange(df.shape[0])

# Bar width
bar_width = 0.2

plt.bar(x-bar_width,df['2015'],width=bar_width, label='2015')
plt.bar(x,df['2016'],width=bar_width, label='2016')
plt.bar(x+bar_width,df['2017'],width=bar_width, label='2017')

#labeling
plt.title('Runs by Year for Each Batsman')
plt.xlabel('Batsman')
plt.ylabel('Runs Scored')

#plt.xticks(x-bar_width,df['batsman'])
plt.xticks(x-bar_width,df['batsman'], rotation=45, ha="right")
plt.legend()
plt.show()
```

## Stack Bar Chart

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
# percentages on pie chart

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode
#shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
