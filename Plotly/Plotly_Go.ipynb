```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.offline as pyo
import plotly.graph_objects as go
```

```python
match = pd.read_csv('/content/batter.csv')
match.head()
```

#Scatterplot

```python
# batsman avg(x-axis) vs strikerate(y-axis) on top 50 batsman

top50 = match.groupby('batter')['runs'].sum().sort_values(ascending=False).head(50).index.to_list()

df = match[match['batter'].isin(top50)]
df.head()
```

```python
# basic structure or hiarerchy

trace = go.scatter(x,y,mode)

data = [trace]
layout = go.Layout()

fig = go.figure(data,layout)

fig.show()
```

```python
#mode/text/marker/

# mode=markers/lines/text
# text = hover korle ki ashbe sheta

trace = go.Scatter(x= df['avg'], y= df['strike_rate'],mode='markers', text=df['batter'],
                   marker ={'color':'red','size':16} )

data = [trace]
layout = go.Layout(title = {'text': 'Batsman Avg vs Strike_rate', 'x': 0.5},
                  xaxis = {'title':'Batsman Average'},
                  yaxis = {'title':'Batsman Strike_rate'})

fig = go.Figure(data=data,layout=layout)

fig.show()
```

# Linechart

```python
df.head()
```

```python
trace = go.Scatter(x= df['batter'], y= df['runs'],mode='lines+markers',
                   marker ={'color':'red','size':10} )

data = [trace]
layout = go.Layout(title = {'text': 'Batsman Avg vs Strike_rate', 'x': 0.5},
                  xaxis = {'title':'Batsman Average'},
                  yaxis = {'title':'Batsman Strike_rate'})

fig = go.Figure(data=data,layout=layout)

fig.show()
```

```python
trace1 = go.Scatter(x= df['batter'], y= df['strike_rate'],mode='lines+markers',
                   marker ={'color':'red','size':10},name='Strike_rate' )

trace2 = go.Scatter(x= df['batter'], y= df['avg'],mode='lines+markers',
                   marker ={'color':'Green','size':10}, name='Average' )

data = [trace1,trace2]
layout = go.Layout(title = {'text': 'Batsman Avg vs Total_run', 'x': 0.5},
                  xaxis = {'title':'Batsmen'},
                  yaxis = {'title':'Batsman Strike_rate and Average'})

fig = go.Figure(data=data,layout=layout)

fig.show()
```

```python
years = ['2016','2017','2018']

fig = go.Figure()
fig.add_trace(go.Bar(x=years, y=[500, 600, 700],
                base=[-500,-600,-700],
                marker_color='crimson',
                name='expenses'))
fig.add_trace(go.Bar(x=years, y=[300, 400, 800],
                base=0,
                marker_color='lightslategrey',
                name='revenue'
                ))

fig.show()
```
